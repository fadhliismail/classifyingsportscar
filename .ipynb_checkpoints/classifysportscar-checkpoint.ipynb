# Classifying sports car

For this project, I decided to built a classifier that able to distinguish between car made by Ferrari, Lamborghini, and Aston Martin. I built the dataset myself using with pictures of cars downloaded through Google images.

```python
from fastai import *
from fastai.vision import *
```

## Preparing the data

```python
path = Path('data/cars')
```

```python
folder = 'ferrari'
file = 'ferrari.txt'
```

```python
dest = path/folder
dest.mkdir(parents=True, exist_ok=True)
```

```python
download_images(path/file, dest, max_pics=200)
```

```python
folder = 'lamborghini'
file = 'lamborghini.txt'
```

```python
dest = path/folder
dest.mkdir(parents=True, exist_ok=True)
```

```python
download_images(path/file, dest, max_pics=200)
```

```python
folder = 'astonmartin'
file = 'astonmartin.txt'
```

```python
dest = path/folder
dest.mkdir(parents=True, exist_ok=True)
```

```python
download_images(path/file, dest, max_pics=200)
```

```python
classes = ['ferrari', 'lamborghini', 'astonmartin']
```

```python
for c in classes:
    print(c)
    verify_images(path/c, delete=True, max_workers=8)
```

## View Data

```python
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".",
                                      valid_pct=0.2,
                                      ds_tfms=get_transforms(),
                                        size=224, num_workers=4).normalize(imagenet_stats)
```

```python
data.classes
```

```python
data.show_batch(rows=5, figsize=(7,8))
```

```python
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
```

```python
ls data/cars/ferrari | wc -l
```

```python
ls data/cars/lamborghini | wc -l
```

```python
ls data/cars/astonmartin | wc -l
```

## Train model

```python
learn = create_cnn(data, models.resnet34, metrics=error_rate)
```

```python
learn.fit_one_cycle(4)
```

```python
learn.save('stage-1')
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(2, max_lr=slice(3e-5, 1e-4))
```

```python
learn.save('stage-2')
```

## Interpretation

```python
learn.load('stage-2')
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(9, figsize=(15,11))
```
